```python

# Basic data types
integer_var = 42
float_var = 3.14
string_var = "Hello, World!"
boolean_var = True

# Collections
list_var = [1, 2, 3, 4, 5]
tuple_var = (10, 20, 30)
dict_var = {"name": "Alice", "age": 30, "city": "New York"}
set_var = {1, 2, 3, 4, 5}

# Display data types
print(f"Type of integer_var: {type(integer_var)}")
print(f"Type of float_var: {type(float_var)}")
print(f"Type of string_var: {type(string_var)}")
print(f"Type of boolean_var: {type(boolean_var)}")
print(f"Type of list_var: {type(list_var)}")
print(f"Type of tuple_var: {type(tuple_var)}")
print(f"Type of dict_var: {type(dict_var)}")
print(f"Type of set_var: {type(set_var)}")

# Display values
print("\nValues:")
print(f"integer_var: {integer_var}")
print(f"float_var: {float_var}")
print(f"string_var: {string_var}")
print(f"boolean_var: {boolean_var}")
print(f"list_var: {list_var}")
print(f"dict_var: {dict_var}")
```

```python
if integer_var > 40:
    print("integer_var is greater than 40")
else:
    print("integer_var is not greater than 40")

if boolean_var:
    print("boolean_var is True")

if "Alice" in dict_var.values():
    print("Alice is in dict_var")
else:
    print("Alice is not in dict_var")
```

```python
if integer_var > 0:
    print("integer_var is positive")
else:
    print("integer_var is not positive")

if "Bob" in dict_var.values():
    print("Bob is in dict_var")
else:
    print("Bob is not in dict_var")

for value in list_var:
    if value < 0:
        print("Found a negative number")
        break
else:
    print("No negative numbers found in list_var")
```

```python
if integer_var > 50:
    print("integer_var is greater than 50")
elif integer_var == 42:
    print("integer_var is exactly 42")
else:
    print("integer_var is less than 50 and not 42")
```

```python
for item in list_var:
    if item % 2 == 0:
        print(f"{item} is even")
        continue
    elif item == 5:
        print(f"{item} is odd and equals 5, breaking")
        break
    else:
        print(f"{item} is odd and less than 5")
else:
    print("Finished loop without breaking")
```

```python
i = 0
while i < len(list_var):
    if list_var[i] == 3:
        print("Skipping value 3")
        i += 1
        continue

    if list_var[i] == 5:
        print("Found 5, breaking out of the loop")
        break

    print(f"Current value: {list_var[i]}")
    i += 1
else:
    print("Finished while loop without breaking")
```

```python
# Python does not have a built-in do-while loop, but we can simulate it
count = 0
while True:
    print(f"Do-while style iteration: count={count}")
    count += 1
    if count >= 5:
        break
```
